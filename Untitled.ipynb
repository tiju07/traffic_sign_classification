```python
#Reshaping to (50, 50)
images = []
label_id = []

for i in range(43):
    labels = 'traffic_sign_dataset/Train' + '/{0}'.format(i)
    image_path = os.listdir(labels)
    for x in image_path:
        img = Image.open(labels + '/' + x)
        img = img.resize((50, 50))
        img = np.array(img)
        images.append(img)
        label_id.append(i)
```
